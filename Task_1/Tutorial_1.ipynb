```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mu,sigma=5,1
x = np.random.normal(mu,sigma,10)
np.mean(x)
```

```python
def get_liklihood(z):
    ans= 1/(np.sqrt(2*np.pi*sigma**2)) * np.exp(-0.5*((x-z)/sigma)**2)
    return ans
```

```python
y = 1/(np.sqrt(2*np.pi*sigma**2)) * np.exp(-0.5*((x-mu)/sigma)**2)
y
```

```python
sns.lineplot(x,y)
plt.savefig("guassian.png")
```

```python
plt.figure(figsize = (20,10))
for i in range(1,12,1):
    y1=get_liklihood(i)
    plt.subplot(3,4,i)
    sns.lineplot(x,y1)
plt.savefig("plot.png")
```
